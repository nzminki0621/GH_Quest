```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

music_db_df = pd.read_csv("./music_db.csv")
music_db_major = music_db_df[music_db_df['mode']=='Major']
music_db_minor = music_db_df[music_db_df['mode']=='Minor']

plt.figure(figsize=(10,8))
plt.plot(music_db_major.loudness,music_db_major.liveness,'o')
plt.plot(music_db_minor.loudness,music_db_minor.liveness,'o',color='red')
plt.show()

music_db_df.corr()['loudness']
```
